скачиваем русский датасет без текста

```python
from datasets import load_dataset

# Загружаем датасет
ds = load_dataset("ZeroAgency/ru-tasks-conversation")

# Смотрим информацию о датасете
print("Доступные сплиты:", ds.keys())
print("Размер тренировочного набора:", len(ds['train']))

# Посмотрим на первый пример
print("\nПример диалога:")
first_example = ds['train'][0]
conversation = first_example['conversation']

print("Структура диалога:")
for message in conversation:
    print(f"\nРоль: {message['role']}")
    print(f"Содержание: {message['content'][:200]}...")  # Показываем первые 200 символов

# Проверим формат данных
print("\nСтруктура данных:")
print(ds['train'].features)
```

инициализируем tts модель

```python
import outetts
import random
import os

def setup_outetts():
    """Инициализация OuteTTS"""
    interface = outetts.Interface(
        config=outetts.ModelConfig.auto_config(
            model=outetts.Models.VERSION_1_0_SIZE_1B,
            backend=outetts.Backend.LLAMACPP,
            quantization=outetts.LlamaCppQuantization.FP16
        )
    )
    return interface

def generate_audio(interface, text, role, output_path):
    """Генерация аудио с разными параметрами для вопросов и ответов"""
    # Выбираем разные голоса для вопросов и ответов
    speaker_id = "RU-FEMALE-1-NEUTRAL" if role == "user" else "RU-MALE-1-NEUTRAL"
    speaker = interface.load_default_speaker(speaker_id)

    # Настраиваем параметры голоса
    config = outetts.GenerationConfig(
        text=text,
        generation_type=outetts.GenerationType.CHUNKED,
        speaker=speaker,
        sampler_config=outetts.SamplerConfig(
            temperature=0.4 if role == "user" else 0.5,  # Больше вариативности для ответов
            repetition_penalty=1.1,
            repetition_range=64,
            top_k=40,
            top_p=0.9,
            min_p=0.05
        ),
        # Рандомизация параметров
        pitch_shift=random.uniform(-0.1, 0.1),
        speed=random.uniform(0.95, 1.05) if role == "user" else random.uniform(0.9, 1.0),  # Ответы чуть медленнее
        energy=random.uniform(0.9, 1.1)
    )

    output = interface.generate(config)
    output.save(output_path)

# Создаем директорию для тестовых аудио
!mkdir -p test_audio

# Инициализируем OuteTTS
try:
    tts_interface = setup_outetts()
    print("OuteTTS успешно инициализирован")

    # Берем первый диалог для теста
    test_dialog = ds['train'][0]['conversation']

    # Генерируем аудио для вопроса и ответа
    for i, message in enumerate(test_dialog):
        output_path = f"test_audio/{message['role']}_{i}.wav"
        print(f"\nГенерация аудио для {message['role']}...")
        print(f"Текст: {message['content'][:100]}...")  # Показываем начало текста

        generate_audio(
            tts_interface,
            message['content'],
            message['role'],
            output_path
        )
        print(f"Аудио сохранено в {output_path}")

except Exception as e:
    print(f"Ошибка при работе с OuteTTS: {str(e)}")
```
